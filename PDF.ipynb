```python
# Standard normal variate
titanic['age'].mean()
```

```python
sns.kdeplot(titanic['age']) # mean is 24
```

```python
titanic['age'].std()
```

```python
x = (titanic['age']-titanic['age'].mean())/titanic['age'].std()
```

```python
sns.kdeplot(x) # mean is '0'
```

```python
# Use of Normal distribution

titanic['age'].mean() + 3*titanic['age'].std() # std value is +3 (for +ve)
```

```python
titanic['age'].mean() - 3*titanic['age'].std()# std=-3
```
